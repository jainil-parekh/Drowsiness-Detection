```python
# Datadirectory = "C:\\Users\\jpare\\Downloads\\Test_Dataset"
# Classes = ["Closed_Eyes","Open_Eyes"] 
# for category in Classes:
#     path = os.path.join(Datadirectory, category)
#     for img in os.listdir(path):
#         img_array = cv2.imread(os.path.join(path,img), cv2.img_size)
#         backtorgb = cv2.cvtColor(img_array,cv2.COLOR_GRAY2RGB)
#         plt.imshow(img_array, cmap="gray")
#         plt.show()
#         break   read and display images from a dataset,
#     break   checks for directory and file existence, displays the first image from the first category found

import os
import cv2
import matplotlib.pyplot as plt

Datadirectory = r"C:\Users\jpare\Downloads\Test_Dataset"
Classes = ["Closed_Eyes", "Open_Eyes"]

for category in Classes:
    path = os.path.join(Datadirectory, category)
    print(f"Checking path: {path}")  # Print the path for debugging
    if not os.path.exists(path):
        print(f"Path does not exist: {path}")
        continue  # Skip this iteration if the path does not exist
    
    for img in os.listdir(path):
        img_path = os.path.join(path, img)
        if not os.path.isfile(img_path):
            print(f"File does not exist: {img_path}")
            continue  # Skip if the file does not exist

        img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img_array is None:
            print(f"Failed to read image: {img_path}")
            continue  # Skip if the image cannot be read

        backtorgb = cv2.cvtColor(img_array, cv2.COLOR_GRAY2RGB)
        plt.imshow(img_array, cmap="gray")
        plt.show()
        break
    break
```

```python
img_size=224
new_array= cv2.resize(backtorgb, (img_size,img_size))
plt.imshow(new_array, cmap="gray")
plt.show()
```

```python
#reading all images and converting them to array for data and labels
training_Data = []
def create_training_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                backtorgb = cv2.cvtColor(img_array,cv2.COLOR_GRAY2RGB)
                new_array= cv2.resize(backtorgb, (img_size,img_size))
                training_Data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_Data()
```

```python
print(len(training_Data))
```

```python
import random
random.shuffle(training_Data)
```

```python
X = []
y = []
for features,label in training_Data:
    X.append(features)
    y.append(label)
X = np.array(X).reshape(-1, img_size, img_size, 3)
```

```python
X.shape
```

```python
#normalize the data
X = X/255.0;
```

```python
Y= np.array(y)
```

```python
import pickle

pickle_out = open("X.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle","wb")
pickle.dump(y, pickle_out)
pickle_out.close()

```

```python
#deep learning model for training 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.mobilenet.MobileNet()
```

```python
model.summary()
```

```python
from tensorflow.keras import layers, models

# Print layer details to verify indices
for i, layer in enumerate(model.layers):
    if hasattr(layer, 'input_shape') and hasattr(layer, 'output_shape'):
        print(i, layer.name, layer.input_shape, layer.output_shape)
    elif hasattr(layer, 'batch_input_shape') and hasattr(layer, 'output_shape'):
        print(i, layer.name, layer.batch_input_shape, layer.output_shape)

# Check and print model inputs
model_inputs = model.inputs
print("Model Inputs:")
for i, inp in enumerate(model_inputs):
    print(f"Input {i}:")
    print("Shape:", inp.shape)
    print("Type:", type(inp))

# Verify base_input and base_output
if len(model_inputs) > 0:
    base_input = model_inputs[0]
else:
    raise ValueError("No inputs found in the model.")

base_output = model.layers[-4].output

print("Base Input Layer Details:")
print("Shape:", base_input.shape)
print("Type:", type(base_input))

print("Base Output Layer Details:")
print("Shape:", base_output.shape)
print("Type:", type(base_output))

# Apply Flatten layer
Flat_layer = layers.Flatten()(base_output)

# Add a Dense layer
final_output = layers.Dense(1)(Flat_layer)

# Add an Activation layer
final_output = layers.Activation('sigmoid')(final_output)

# Create the new model
try:
    new_model = models.Model(inputs=base_input, outputs=final_output)
    # Print the model summary to verify
    new_model.summary()
except Exception as e:
    print("Error creating new model or printing summary:", e)
```

```python
new_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
new_model.fit(X,Y, epochs = 1,validation_split = 0.1)  #training
```

```python
new_model.save("my_model.keras")
```

```python
import cv2
import numpy as np

# Load the cascades
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eyeCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Attempt to open the camera
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret, frame = cap.read()
    if not ret or frame is None:
        print("Failed to capture image")
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]

        # Detect eyes
        eyes = eyeCascade.detectMultiScale(roi_gray, 1.1, 4)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
            eyes_roi = roi_color[ey:ey+eh, ex:ex+ew]

            try:
                final_image = cv2.resize(eyes_roi, (224, 224))
                final_image = np.expand_dims(final_image, axis=0)  # Need 4th dimension
                final_image = final_image / 255.0

                # Assuming 'new_model' is already defined and loaded somewhere
                Predictions = new_model.predict(final_image)
                if Predictions > 0:
                    status = "Open Eyes"
                else:
                    status = "Closed Eyes"
            except Exception as e:
                status = "Error: " + str(e)

            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(frame, status, (50, 50), font, 1, (0, 0, 255), 2, cv2.LINE_4)

    # Show the frame
    cv2.imshow("Drowsiness Detection Tutorial", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
